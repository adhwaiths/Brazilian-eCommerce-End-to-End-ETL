```python
def data_load(file_name,file_schema,table_name,mode):
  try:
    catalog="Brazilian_Ecommerce"
    schema="Landing"
    bronze_df=spark.read.format("csv").option("header","true").schema(file_schema).load(f"/Volumes/{catalog}/{schema}/landing_zone/{file_name}")
    # Check if the file contains data or not.
    if len(bronze_df.take(1)) == 0:
      raise Exception(f"File {file_name} contains no data.")
    bronze_df.write.mode(mode).saveAsTable(f"{catalog}.bronze.{table_name}")
    return "Table successfully created"
  except Exception as e:
    raise Exception(f"Error: {e}")
  

def bronze_ingestion():
  try:
    file_schema=f"""customer_id string,customer_zip_code_prefix int,customer_city string,customer_state string"""
    file_name='olist_customers_dataset.csv'
    table_name='customers'
    data_load(file_name,file_schema,table_name,"overwrite")
    
    
    file_schema="product_id string,product_category_name string,product_name_length int,product_description_length int,product_photos_qty int,product_weight_g int,product_length_cm int,product_height_cm int,product_width_cm int"
    file_name='olist_products_dataset.csv'
    table_name='Products'
    data_load(file_name,file_schema,table_name,"overwrite")
    
    
    file_schema="order_item_id int,order_id string,customer_id string,order_status string,order_purchase_timestamp timestamp,order_approved_at timestamp,order_delivered_carrier_date timestamp,order_delivered_customer_date timestamp,order_estimated_delivery_date timestamp,product_id string,seller_id string,payment_type string,price float,freight_value float,payment_installments int,payment_value float"
    file_name='olist_orders_dataset.csv'
    table_name='Orders'
    data_load(file_name,file_schema,table_name,"append")
    
    file_schema="seller_id string,seller_zip_code_prefix int,seller_city string,seller_state string"
    file_name='olist_sellers_dataset.csv'
    table_name='Sellers'
    data_load(file_name,file_schema,table_name,"overwrite")
    
    
    file_schema="product_category_name string,product_category_name_english string"
    file_name='product_category_name_translation.csv'
    table_name='Product_Category'
    data_load(file_name,file_schema,table_name,"overwrite")
    
    print("Bronze ingestion was completed successfully")
  except Exception as e:
    raise Exception(f"Error: {e}")


if __name__ == "__main__":
    print("Bronze Ingestion Started")
    bronze_ingestion()
    print("Bronze Ingestion Completed")
```
